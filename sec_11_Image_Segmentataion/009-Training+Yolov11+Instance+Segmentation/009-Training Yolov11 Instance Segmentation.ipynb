## Files information validation

```python
import os
```

```python
def get_files(folder_path):
  if os.path.exists(folder_path) and os.path.isdir(folder_path):
      files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]
      num_files = len(files)
      print(f"Number of files in '{folder_path}': {num_files}")
  else:
      print(f"Folder '{folder_path}' does not exist or is not a directory.")
```

```python
def check_files(dir_path, structure=['images', 'labels']):
  folders = os.listdir(dir_path)
  for folder in folders:
    folder_path = os.path.join(dir_path, folder)
    if os.path.isdir(folder_path):
      sub_folders = os.listdir(folder_path)
      for final_folders in sub_folders:
        if final_folders in structure:
          get_files(os.path.join(folder_path, final_folders))
```

```python
check_files('/content/dataset')
```

```python
import zipfile
import os

def zip_folder(folder_path, output_zip_path):
    try:
        with zipfile.ZipFile(output_zip_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
            for root, dirs, files in os.walk(folder_path):
                for file in files:
                    file_path = os.path.join(root, file)
                    arc_name = os.path.relpath(file_path, folder_path) # get relative path
                    zipf.write(file_path, arc_name)
        print(f"Folder '{folder_path}' zipped successfully to '{output_zip_path}'")
    except FileNotFoundError:
        print(f"Error: Folder '{folder_path}' not found.")
    except Exception as e:
        print(f"An error occurred: {e}")
```

```python
# Example usage:
folder_to_zip = "/content/Yolov11-test-1"  # Replace with the path to your folder
output_zip_file = "/content/Yolov11-test-1.zip"  # Replace with the desired output zip file path

zip_folder(folder_to_zip, output_zip_file)
```
